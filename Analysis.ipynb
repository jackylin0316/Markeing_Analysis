## Task#3 :Which age group purchased the most?

```python
df.columns
```

```python
task=df[['ID', 'Year_Birth', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases']]
task['Total_Purchase']=task.iloc[:, 2:].sum(axis=1)
task
```

```python
# child : 0-16 
# young adults : 17-30
# middle-aged adults : 31-45
# old-aged adults : above 45    
def group(x):
    if x>=2005:
        return 'child'
    elif x<2005 and x>=1991:
        return 'young adult'
    elif x>=1976 and x<=1990:
        return 'middle-aged adults'
    elif x<1976:
        return 'senior'
    
task['age_group'] = task['Year_Birth'].apply(lambda x: group(x))
task
```

```python
Count=task.groupby('age_group').count()['ID']
```

```python
Total_purchase=task.groupby('age_group').sum()['Total_Purchase']
```

```python
task=pd.concat([Count, Total_purchase], axis=1)
```

```python
task=task.reset_index()
task
```

```python
task = task.rename(columns={'ID': 'Count'})
task
```

```python
x = task['age_group']
#y = [task.loc['Total', 'AcceptedCmp1'], task.loc['Total', 'AcceptedCmp2'], task.loc['Total', 'AcceptedCmp3'], task.loc['Total', 'AcceptedCmp4']]
y=task['Total_Purchase']
fig, ax = plt.subplots(figsize=(8, 8))    
width = 0.2 # the width of the bars 
ind = np.arange(len(x))  # the x locations for the groups
ax.bar(ind, y, width, color="red")
ax.set_xticks(ind)
ax.set_xticklabels(x, fontsize=12,minor=False)
plt.xlabel('age_group', fontsize=12)
plt.ylabel('how many times they bought a merchandise', fontsize=12)    
for i, v in enumerate(y):
    ax.text(i-0.09, v+500, str(v), color='blue', fontweight='bold')
plt.show()
```

## Which product was sold the most in all campagins?

```python
df.head()
```

```python
task=df.iloc[:, 9:15]
task.head()
```

```python
task.loc['Total']=task.sum(axis=0)
task
```

```python
import re
p=pd.DataFrame({'quantity':task.loc['Total', :]})
p=p.reset_index()
p=p.rename(columns={'index':'product'})
def rem(x):
    x = re.sub('Mnt', '', x)
    return x
p['product']=p['product'].apply(lambda x: rem(x))
p
```

```python
x = p['product']
y=p['quantity']
fig, ax = plt.subplots(figsize=(8, 8))    
width = 0.2 # the width of the bars 
ind = np.arange(len(x))  # the x locations for the groups
ax.bar(ind, y, width, color="red")
ax.set_xticks(ind)
ax.set_xticklabels(x, fontsize=12,minor=False, rotation=90)
plt.xlabel('Product', fontsize=12)
plt.ylabel('sold quantity', fontsize=12)    
for i, v in enumerate(y):
    ax.text(i-0.2, v+10000, str(v), color='blue', fontweight='bold')
plt.show()
```

## So the answer is Wines
